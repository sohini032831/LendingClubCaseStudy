```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv("C:/Users/sohbanerjee/Desktop/Loan Data Set.csv", low_memory=False)
data.head()
```

```python
#Dropping customer behavior variables
data.drop(["delinq_2yrs"],axis = 1, inplace = True)
```

```python
data
data.drop(["revol_bal","revol_util", "total_acc"],axis = 1, inplace = True)
```

```python
data.drop(["earliest_cr_line","inq_last_6mths"], axis = 1, inplace=True)
```

```python
data.drop(["open_acc","pub_rec","out_prncp","out_prncp_inv"], axis = 1, inplace=True)
```

```python
data.drop(["total_pymnt","total_pymnt_inv","total_rec_prncp","total_rec_int","total_rec_late_fee","recoveries","collection_recovery_fee"],axis=1,inplace=True)
```

```python
data.drop(["last_pymnt_d"], axis = 1, inplace=True)
```

```python
data.drop(["last_credit_pull_d"], axis = 1, inplace=True)
```

```python
data.drop(["application_type"], axis = 1, inplace=True)
```

```python
data.drop(["last_pymnt_amnt"], axis = 1, inplace=True)
```

```python
data
```

```python
#Finding percentage of null or missing values from columns
perc_missing = round(100*(data.isnull().sum()/len(data.index)), 2)
```

```python
#Removing columns which have more than 30% missing values in those
data.drop(perc_missing[ perc_missing > 30 ].index, axis=1, inplace=True)
```

```python
data
```

```python
#Columns "emp_title" and "title" have text values which will not impart any additionl value to the analysis. Hence, removing these columns.
data.drop(["emp_title", "title"], axis=1, inplace=True)
```

```python
#Columns "id","url" and "member_id" all have unique values. Hence, we will use "id" as the primary key and remove the other 2 columns.
data.drop(["member_id", "url"], axis=1, inplace=True)
```

```python
data
```

```python
#Finding rows with missing values
perc_missing_row=round(100*(data.isnull().sum()/len(data.index)), 2)
perc_missing_row[ perc_missing_row > 0]
```

```python
#Removing rows with missing values
data.dropna(subset=["emp_length","collections_12_mths_ex_med","chargeoff_within_12_mths","pub_rec_bankruptcies","tax_liens"],inplace=True)
```

```python
#The records for which the “loan_status” is marked as “current” are neither fully paid nor defaulted. Hence, these records should be removed before analyzing the dataset further.
data = data[data.loan_status != 'Current']
```

```python
var = 'term'
#Probability / Percentage of each values
prob_df = data[var].value_counts(normalize=True).reset_index()

sns.barplot(x='index', y=var, data=prob_df)
plt.xlabel(var)
plt.ylabel('Proportion')
plt.title(var+' Distribution')
plt.show()
```

```python
plt.figure(figsize=(10,4))
var = 'loan_status'
#Probability / Percentage of each values
prob_df = data[var].value_counts(normalize=True).reset_index()

sns.barplot(x='index', y=var, data=prob_df)
plt.xlabel(var)
plt.ylabel('Proportion')
plt.title(var+' Distribution')
plt.show()
```

```python
#Term vs Loan Status
df = data.groupby(['term', 'loan_status'], as_index=False)['id'].count()
df['proportion'] = df.groupby('term').transform(lambda x: x/x.sum())
sns.barplot(x='term', y='proportion', hue='loan_status', data=df, hue_order = ['Fully Paid', 'Charged Off'])

plt.title('Term vs Loan status')
```

```python
#Term vs Loan Amount
plt.figure(figsize=(20,5))
plt.subplot(121)
sns.boxplot(x='term', y='loan_amnt', hue='loan_status', data=data)
plt.title('Term vs loan amount')
plt.subplot(122)
sns.barplot(x='term', y='loan_amnt', hue='loan_status', data=data, estimator=np.median)
plt.title('Term vs loan amount')
plt.show()
```

```python
#Term vs DTI
plt.figure(figsize=(20,5))
plt.subplot(121)
sns.boxplot(x='term', y='dti', hue='loan_status', data=data)
plt.subplot(122)
sns.barplot(x='term', y='dti', hue='loan_status', data=data, estimator=np.median)
plt.suptitle('Term vs DTI')
plt.show()
```

```python
#Grade vs Loan Status
df = data.groupby(['grade', 'loan_status'], as_index=False)['id'].count()
df['percentage'] = df.groupby('grade').transform(lambda x: x/x.sum())
plt.figure(figsize=(12,5))
sns.barplot(x='grade', y='percentage', hue='loan_status', data=df, hue_order = ['Fully Paid', 'Charged Off'])
plt.title('Grade vs Loan status')
```

```python
#Employment Length vs Loan Status
df = data.groupby(['emp_length', 'loan_status'], as_index=False)['id'].count()
df['percentage'] = df.groupby('emp_length').transform(lambda x: x/x.sum())
plt.figure(figsize=(12,5))
sns.barplot(x='emp_length', y='percentage', hue='loan_status', data=df)
plt.title('Employment Length vs Loan status')
```

```python
#State vs Charged Off
charged_off_df = data[data.loan_status.values == 'Charged Off']
plt.figure(figsize=(20,6))
plt.subplot(211)
sns.countplot(x='addr_state', data=charged_off_df, order=charged_off_df.addr_state.value_counts().index)
plt.title('State vs Charged Off')
```

```python
#Purpose vs Loan Status
df = data.groupby(['purpose', 'loan_status'], as_index=False)['id'].count()
df['percentage'] = df.groupby('purpose').transform(lambda x: x/x.sum())
df = df[df.loan_status == 'Charged Off']
plt.figure(figsize=(12,5))
sns.barplot(x='purpose', y='percentage', data=df)
plt.xticks(rotation=90)
plt.title('Purpose vs Loan status (Charged Off)')
```
